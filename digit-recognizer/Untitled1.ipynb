```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.utils import np_utils
from keras.layers.core import Dense, Activation, Dropout
import pandas as pd
import numpy as np
```

```python
#get feature columns
data=pd.read_csv('train.csv')
columns=data.drop('label',axis=1).columns
feature_col=[]
for x in columns:
    feature_col.append(tf.feature_column.numeric_column(key=x))
    
print(feature_col)  
```

```python
from tensorflow.estimator import DNNClassifier
# Or estimator using an optimizer with a learning rate decay.
estimator = DNNClassifier(feature_columns=feature_col,hidden_units=[784,10],n_classes=10)
```

```python
def input_fn_train():
    data=pd.read_csv('train.csv')
    cols=data.drop('label',axis=1).columns
    feature={}
    for x in cols:
        feature[x]=data[x]
    return feature,data['label']
    
estimator.train(input_fn=input_fn_train)


```

```python
def input_fn_predict():
    data=pd.read_csv('test.csv')
    cols=data.columns
    feature={}
    for x in cols:
        feature[x]=data[x]
    return feature

predictions=estimator.predict(input_fn=input_fn_predict)
```
